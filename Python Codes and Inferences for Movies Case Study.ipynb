```python
#Import pandas
import pandas as pd
```

```python
#Get Raw data file users.dat into a data frame called users
#Make sure to name columns in the order specified in the raw data set as observed in a file preview in the Jupter Notebook
#Note the delimiter in the file (::) as observed in a file preview in the Jupyter notebook
user_columns = ['User_ID', 'Gender', 'Age', 'Occupation', 'Zip_Code']
users = pd.read_csv('users.dat', sep='::', names=user_columns)
```

```python
#Get a preview of the 1st 5 rows of the users data frame
users.head()
```

```python
#The Following age categories have been created for the data frame called users
#Young: <18 years
#Young Adult: 18 - 44 years
#Adult > 44 years
#Recall that the ages of users in the data set have been categorized in the following way:
    #1:  "Under 18"
    #18:  "18-24"
    #25:  "25-34"
    #35:  "35-44"
    #45:  "45-49"
    #50:  "50-55"
    #56:  "56+"
```

```python
def age_cat4(users):
    age,user_id=users
    if age in range(18,36):
        age_cat='Young Adult'
    elif age in range(45,57):
        age_cat='Adult'
    else:
        age_cat='Young'
    return age_cat        
```

```python
users['age_cat'] = users[['Age','User_ID']].apply(age_cat4,axis=1)
```

```python
users.head()
```

```python
#Get Raw data file movies.dat into a data frame called movies
#Make sure to name columns in the order specified in the raw data set as observed in a file preview in the Jupter Notebook
#Note the delimiter in the file (::) as observed in a file preview in the Jupyter notebook
movie_columns = ['Movie_ID', 'Title', 'Genres']
movies = pd.read_csv('movies.dat', sep='::', names=movie_columns)
```

```python
#Get a preview of the 1st 5 rows of the movies data frame
movies.head()
```

```python
#Get Raw data file users.dat into a data frame called ratings
#Make sure to name columns in the order specified in the raw data set as observed in a file preview in the Jupter Notebook
#Note the delimiter in the file (::) as observed in a file preview in the Jupyter notebook
rating_columns = ['User_ID', 'Movie_ID', 'Rating', 'Timestamp']
ratings = pd.read_csv('ratings.dat', sep='::', names=rating_columns)
```

```python
#Get a preview of the 1st 5 rows of the ratings data frame
ratings.head()
```

```python
#Merge all the three data frames together
#First merge the movies and ratings data frames
#The common column is Movie_ID
movies_and_ratings = pd.merge(movies, ratings)
```

```python
#Get a preview
movies_and_ratings.head()
```

```python
#Merge movies_and_ratings with users
#The common column is user_id
final_dataframe = pd.merge(movies_and_ratings, users)
```

```python
#Get a preview of the first five rows:
final_dataframe.head()
```

```python
#Question 1: Top ten most viewed movies with their movies Name
#Set up a new data frame and group it by the title of the movies
#Use the size() function to obtain a count the number of user records associated 
#with each movie - each movie constitutes a separate group here
#Make sure to input false for the ascending = option - we want the values sorted from largest to smallest
#I have made the assumption of associating the count of the number of user records for each movie as the
#number of views received for that specific movie
Grouped = pd.DataFrame({'Views':final_dataframe.groupby('Title').size().sort_values(ascending = False)}).reset_index()
```

```python
#Preview of Grouped
Grouped.head()
```

```python
#Name a new data frame called Top_Ten with only the Top 10 most viewed movies
Top_Ten = Grouped.head(10)
```

```python
Top_Ten
```

```python
#Question 2: Top twenty rated movies (Condition : The movie should be rated/viewed by at least 40 users)
```

```python
#Create another data frame called avg_ratings
#First group final_dataframe by the Movie Title
#Named the grouped data frame as Grouped2
Grouped2 = final_dataframe.groupby('Title')
```

```python
#Extract the average ratings for each movie from the data frame Grouped2
#using the mean() method
avg_ratings = Grouped2['Rating'].mean()
```

```python
#Get a preview
#Note: avg_ratings is a Series Object
avg_ratings.head()
```

```python
#Again, group final_dataframe by the Movie Title
#Named the grouped data frame as Grouped3
Grouped3 = final_dataframe.groupby('Title')
```

```python
#Extract the count (number of user views) for each movie from the data frame Grouped2
#using the method size()
user_counts = Grouped3['Title'].size()
```

```python
#Get a preview
#Note: user_counts is a Series Object
user_counts.head()
```

```python
#Join the two series user_counts and avg_ratings
#Concatenate along column axis
#Call this data frame concatenated
concatenated = pd.concat([user_counts, avg_ratings], axis=1)
```

```python
concatenated.head()
```

```python
#Clean up the data frame
#Rename column 1 as user_counts in concatenated data frame
concatenated.rename(columns={0 : 'User_Counts'}, inplace = True)
```

```python
#Get a preview of the data frame
concatenated.head()
```

```python
#Selecting a subset of the observations in the data frame
#Create another data frame called atleast_or_more_40 which will consist of
#all those movies that have user_counts (user views) of 40 or more
atleast_or_more_40 = concatenated[concatenated['User_Counts']>=40]
```

```python
#Preview of data set atleast_or_more_40
atleast_or_more_40.head()
```

```python
#Create a new data frame called atleast_40
#With a new column called Average_Rating where the average rating for all those movies
#that have 40 or more user views, sorted in descending order from highest to lowest (that's why ascending=False)
#Here, the data frame atleast_or_more is reset into a data frame (I did not want atleast_40 to be a Series object), 
#where Title has been set as columns using reset_index() - without reset_index() - Title will be an index in the 
#data frame atleast_40
atleast_40 = pd.DataFrame({'Average_Rating': atleast_or_more_40['Rating'].sort_values(ascending=False)}).reset_index()
```

```python
#Highest_rated is a data frame
#The first 20 values of the data frame atleast_40 are set equal to the
#highest_rated_20 data frame, using the head(n) method, where here n=20
highest_rated_20 = atleast_40.head(20)
```

```python
highest_rated_20
```

```python
#Question 3: Top twenty rated movies with number of views for each of the age groups
##The Following age categories have been created for the data frame called users
#Young: <18 years
#Young Adult: 18 - 44 years
#Adult > 44 years
#Recall that the ages of users in the data set have been categorized in the following way:
    #1:  "Under 18"
    #18:  "18-24"
    #25:  "25-34"
    #35:  "35-44"
    #45:  "45-49"
    #50:  "50-55"
    #56:  "56+"
```

```python
#First group the original data frame final_dataframe by two factors: age_cat (age category) and Title(Movie Title)
#and obtain the number of user views for each age group (the number of records corresponding to each 
#age category and Movie Title) utilizing the size() method - a new column called Views_By_Age has been created 
#for the data frame Grouped4
#For convenience, age_cat and Title have been reset as columns using the reset_index() method
Grouped4 = pd.DataFrame({'Views_By_Age': final_dataframe.groupby(['age_cat', 'Title']).size()}).reset_index()
```

```python
#Get a preview of the data frame Grouped4
Grouped4.head()
```

```python
#Recall the Data Frame that contains the Top 20 movies
#based on highest average rating and user viewership of 40 or more
highest_rated_20.head()
```

```python
#Convert Title column of the data frame highest_rated_20 to a list
#using the tolist() method
#This list is called Title_List
Title_List = highest_rated_20['Title'].tolist()
```

```python
#Examine the list
Title_List
```

```python
#Subset the data set Grouped4 according to items in the list called Title_List
#Create a new data set called Top_20_By_Age, which is a subset of the data frame Grouped4
#using the isin() method to retain the rows that have Movie Titles contained in the list Title_List
Top_20_By_Age = Grouped4[Grouped4['Title'].isin(Title_List)]
```

```python
#Examine the entire data frame Top_20_By_Age
Top_20_By_Age
```

```python
#Re-shape the data frame Top_20_By_Age into a pivot table and set it equal to a new data frame called Top_20_Age_Views_Final
Top_20_Age_Views_Final = Top_20_By_Age.pivot(index='Title', columns='age_cat', values='Views_By_Age')
```

```python
#Examine the data frame Top_20_Age_Views_Final
Top_20_Age_Views_Final
```

```python
#Question 4: Top Ten Critics - based on the condition that these users have rated at least 40 movies
#Low ratings have been categorized by ratings lower than 2.5 on a 5 point ratings scale
#Group By User ID - create a new data frame Grouped5
Grouped5 = final_dataframe.groupby('User_ID')
```

```python
#Get a preview of Grouped5
Grouped5.head()
```

```python
#Extract Average ratings by each user
#avg_ratings4 is a series object that contains the average of all the ratings assigned, corresponding to each user 
avg_ratings4 = Grouped5['Rating'].mean()
```

```python
#Get a preview of the series object
avg_ratings4.head()
```

```python
#How many movies has each user rated?
#Get the count for each user (in terms of the number of movies that each user has rated)
#using the size() method - set this equal to user_counts4
#user_counts4 is a Series Object
user_counts4 = Grouped5['User_ID'].size()
```

```python
#Get a preview of the Series Object user_counts4
user_counts4.head()
```

```python
#Join the two series user_counts4 and avg_ratings4
#Concatenate along column axis
#Set this equal to a data frame called concatenated2
concatenated2 = pd.concat([user_counts4, avg_ratings4], axis=1)
```

```python
#Preview
concatenated2.head()
```

```python
#Clean up the data frame concatenated2
#Rename column 1 as Number_of_movies_rated in concatenated data frame
concatenated2.rename(columns={0 : 'Number_of_movies_rated'}, inplace = True)
```

```python
#Rename the ratings column as Average of All Ratings Assigned By User
concatenated2.rename(columns={'Rating' : 'Average of All Ratings Assigned By User'}, inplace = True)
```

```python
#Check
concatenated2.head()
```

```python
#Selecting a subset of the observations in the data frame concatenated2
#based on the condition that the Number of Movies rated by a user is 40 or more
#Create another data frame called atleast_40_movies_rated
atleast_40_movies_rated = concatenated2[concatenated2['Number_of_movies_rated']>=40]
```

```python
atleast_40_movies_rated.head()
```

```python
#Selecting a subset of the observations in the data frame atleast_40_movies_rated
#Create another data frame called low_ratings
#Here, we are selecting users whose average of all ratings is less than 2.5 on a scale of 5
low_ratings = atleast_40_movies_rated[atleast_40_movies_rated['Average of All Ratings Assigned By User']<2.5]
```

```python
#Preview of the low_ratings data frame
low_ratings.head()
```

```python
#Create a dataframe called top_10_critics and select the 10 top rated movies using head()
#The column Average_of_Assigned_Ratings is sorted in ascending order (smallest to largest)
#The first 10 observations have been selected using head(n) method where n=10
#The reset_index() method was used to reset the User_ID index and convert it into a column label
top_10_critics = pd.DataFrame({'Average_of_Assigned_Ratings':low_ratings['Average of All Ratings Assigned By User'].sort_values(ascending=True).head(10)}).reset_index()
```
